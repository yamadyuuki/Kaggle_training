```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameter
N = 1024
fs = 100 #サンプリング周波数[Hz]
dt = 1/fs #サンプリング周期
```

```python
#data
import random
t = np.arange(0, N*dt,dt) #時間
f1 = 5 #変動の周期
f2 = 10
f3 = 20


y = np.sin(2* np.pi * f1 * t) + np.sin(2* np.pi * f2 * t) + np.sin(2* np.pi * f3 * t) + 0.5 * np.random.randn(N)
```

```python
plt.plot(t,y)
```

```python
plt.plot(t,y)
plt.xlim(0,1)
```

```python
#FFT
F = np.fft.fft(y)
```

```python
freq = np.fft.fftfreq(N, d = 1/fs)
```

```python
#振幅スペクトル
Amp = np.abs(F)
```

```python
plt.plot(freq[1:int(N/2)], Amp[1:int(N/2)])
plt.xlabel("Frequency [Hz]")
plt.ylabel("Amplitude [#]")
plt.title("FFT test")
```
