```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
plt.hist(df["fare"])
plt.show()
```

```python
plt.hist(df["fare"], bins=40, range=(0,100))
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df["fare"])
axes[0].set_title("Default bins")

axes[1].hist(df["fare"], bins=40, range=(0,100))
axes[1].set_title("bins=40, range=0-100")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df["fare"])
plt.title("titanic data")
plt.xlabel("data_index")
plt.ylabel("fare")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df["fare"].hist(by=df["sex"])
plt.show()
```

```python
# 複数のグラフ描画
fig, axes = plt.subplots(2,2,figsize=(20,10))
axes[0][0].hist(df["age"],bins=20)
axes[0][1].hist(df["fare"],bins=40)
axes[0][1].set_xlim(0,200)
axes[1][0].hist(df["sex"],bins=20)
axes[1][1].hist(df["pclass"],bins=20)
plt.show()
```

```python
print(axes)
```

```python
df.groupby("pclass").count()["survived"]
```

```python
plt.bar(df.groupby("pclass").count()["survived"].index, df.groupby("pclass").count()["survived"])
plt.show()
```

```python
pclass_counts = df.groupby("pclass").count()["survived"]
plt.bar(pclass_counts.index, pclass_counts.values)

plt.xticks([1, 2, 3])  # ← これで横軸を綺麗に固定
plt.xlabel("pclass")
plt.ylabel("count")

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["fare"], df["age"])
plt.xlabel("fare")
plt.ylabel("age")
plt.show()
```
