```python
import numpy as np
```

```python
a = np.array([1,2,3,4,5])
b = np.array([10,20,30,40,50])
#ベクトル演算
print(a + b)
print(a * 2)
```

```python
#リストのパターン
a_list = [1,2,3,4,5]
b_list = [10,20,30,40,50]
print(a_list + b_list)
```

```python
a[1]
a[:3]
```

```python
#統計量計算
a = np.array([1,2,3,4,5])
print(np.sum(a))
print(np.mean(a))
print(np.median(a))
print(np.std(a))
```
