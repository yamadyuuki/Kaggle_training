```python
#Submitした各ファイルを読み込む
import pandas as pd

submit1 = pd.read_csv('/kaggle/input/ansemble-logistic/logisticRegression_submit.csv')
submit2 = pd.read_csv('/kaggle/input/ansemble/kfold_submit (1).csv')
submit3 = pd.read_csv('/kaggle/input/ansemble/lgbm_submission.csv')
```

```python
submit1.head()
```

```python
#各値を確認
submit_all = pd.DataFrame({'submit1': submit1['Survived'].values,
                           'submit2': submit2['Survived'].values,
                           'submit3': submit3['Survived'].values})

submit_all.head(30)
```

```python
#相関を確認
submit_all.corr()
```

```python
#3つの結果を加算
submit_ansemble = pd.read_csv('/kaggle/input/titanic/gender_submission.csv')
submit_ansemble['Survived'] = submit1['Survived'] + submit2['Survived'] + submit3['Survived']

submit_ansemble.head()
```

```python
#結果が2以上なら1とみなす
submit_ansemble['Survived'] = (submit_ansemble['Survived'] >= 2).astype(int)
submit_ansemble.to_csv('ansemble_lightgbm_logistic_submit.csv', index=False)
```
